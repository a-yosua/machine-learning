# Hello world!

This section will build a simple neural network model to learn a relationship between the numbers. We will give the data for the model to learn, containing the relationship between the numbers in set X and Y. This exercise is based on the exercises from "[Introduction to TensorFlow for Artificial Intelligence, Machine Learning, and Deep Learning](https://www.coursera.org/learn/introduction-tensorflow)" course on Coursera.

The model is supposed to predict a value from a number that the model never sees before. The model predicts the value based on the learned relationship from our learning data.

In detail, the model will learn the simple sequence that y = 2x + 3 so for x = 3, y = 9. The following are the function to generate the sequence.

```
def calculate(x):
    return 2 * x + 3

xs = [i * 1.0 for i in range(10)]
ys = []
for i in range(10):
    ys.append(calculate(i) * 1.0)

print(xs)
print(ys)
print('For x is', xs[3], 'y is', ys[3])
```

Next, we will import TensorFlow and define a neural network. The neural network has only 1 layer with 1 neuron. The input shape is 1 value.

The optimizer guesses the number by using the 'Stochastic Gradient Descent', and measures how accurate the model answers the correct answers by using the 'Mean Squared Error'. The training will be repeated 500 times.

```
import tensorflow as tf
import numpy as np
from tensorflow import keras
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
model.fit(xs, ys, epochs=500)
```

Now, our neural network model has been trained to understand the relationship between the numbers in the set X and Y. In the following code, we ask the model to predict a number `y` from the unknown number `x` (that does not exist in our training data). 

The model predicts a number that is very close to the correct answer.

```
print('Given the number x:', 11)
print('The correct answer is:', calculate(11))
print('Our trained model predicts the number y:', model.predict([11.0]))
```
